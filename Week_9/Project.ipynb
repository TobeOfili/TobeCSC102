```python
class Zenith:
    def unique_services(self):
        return

    def mutual_services(self):
        return [
            "Lines of credit",
            "Investment management and accounts",
            "Insurance"]

class RetailBanking(Zenith):
    def unique_services(self):
        return [
            "Lines of credit",
            "Investment management and accounts",
            "Insurance",
            "Retirement and education accounts",
            "Loans and mortgages",
            "Checking and savings"]

class GlobalBanking(Zenith):
    def unique_services(self):
        return [
            "Multi-currency management services and products",
            "Foreign currency accounts",
            "Foreign currency credit cards",
            "Transborder advisory services",
            "Liquidity management"]

class CommercialBanking(Zenith):
    def unique_services(self):
        return [
            "Lines of credit",
            "Investment management and accounts",
            "Insurance",
            "Advisory services"]


def info(name,division):
    if division.lower() == "retail":
        emp_division = RetailBanking()
        print (f"\nThe services for the employee {name.capitalize()} in the {division.capitalize()} banking division are : ")
        display_services(emp_division)
        
    elif division.lower() == "global":
        emp_division = GlobalBanking()
        print (f"\nThe services for the employee {name.capitalize()} in the {division.capitalize()} banking division are : ")
        display_services(emp_division)
    elif division.lower() == "commercial":
        emp_division = CommercialBanking()
        print (f"\nThe services for the employee {name.capitalize()} in the {division.capitalize()} banking division are : ")
        display_services(emp_division)
    else:
        print("Invalid division entered.")


def display_services(emp_division):
    print("Unique services:")
    for service in emp_division.unique_services():
        print("  " + service)

    print("\nMutual services:")
    for service in emp_division.mutual_services():
        print("  " + service)



name = input("Enter employee name: ")
division = input("Enter employee division of banking : ")
info(name, division)
```
