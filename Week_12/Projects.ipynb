```python
from abc import ABC, abstractmethod
import pandas as pd

class External_Vendors(ABC):
    @abstractmethod
    def menu(self):
        pass

class Coop(External_Vendors):
    def menu(self):
        print("\nCooperative Cafeteria Menu:\n")

        print("Main Meal Price        (N)")
        print("Jollof Rice and Stew   200")
        print("White Rice and Stew    200")
        print("Fried Rice             200")
        print("Salad                  100")
        print("Plantain               100")

        Coop_menu = {
            "Main meal": ["Jollof Rice and Stew", "White Rice and Stew", "Fried Rice", "Salad", "Plantain"],
            "Price (N)": [200, 200, 200, 100, 100]
        }

        file = pd.DataFrame(Coop_menu)

        file.to_csv('Cooperative_Menu.csv')
        
        print("\nCooperative_Menu file was created")
class Faith(External_Vendors):
    def menu(self):
        print("\nFaith Hostel Cafeteria Menu:\n")

        print("Main Meal Price        (N)")
        print("Fried Rice             400")
        print("White Rice and Stew    400")
        print("Jollof Rice            400")
        print("Beans                  200")
        print("Chicken                1000")

        Faith_menu = {
            "Main meal": ["Fried Rice", "White Rice and Stew", "Jollof Rice", "Beans", "Chicken"],
            "Price (N)": [400, 400, 400, 200, 1000]
        }
         
        file = pd.DataFrame(Faith_menu)

        file.to_csv('Faith_Menu.csv')
        
        print("\nFaith_Menu file was created")
class StudentCentre(External_Vendors):
    def menu(self):
        print("\nStudent Centre Cafeteria Menu:\n")

        print("Main Meal Price        (N)")
        print("Chicken Fried Rice     800")
        print("Pomo Sauce             300")
        print("Spaghetti Jollof       500")
        print("Amala/Ewedu            500")
        print("Semo with Eforiro Soup 500")

        StudentCentre_menu = {
            "Main meal": ["Chicken Fried Rice", "Pomo Sauce", "Spaghetti Jollof", "Amala/Ewedu", "Semo with Eforiro Soup"],
            "Price (N)": [800, 300, 500, 500, 500]
        }
         
        file = pd.DataFrame(StudentCentre_menu)

        file.to_csv('Student_Centre_Menu.csv')
        
        print("\nStudent_Centre_Menu file was created")
vendor = input("Enter the vendor of interest:\n1.) Cooperative\n2.) Faith Hostel\n3.) Student Centre\n")

if vendor == "Cooperative":
    caf = Coop()
    caf.menu()
elif vendor == "Faith":
    caf = Faith()
    caf.menu()
elif vendor == "Student Centre":
    caf = StudentCentre()
    caf.menu()
else:
    print("Invalid vendor number, try again.")
```
